```python
# Definindo o período (últimos 5 anos)
# Definindo o período (últimos 5 anos)
data_final = dt.date.today()
data_inicial = dt.date(2020, 1, 1)
```

```python
# 1. Coletando a série histórica do CDI (Código 12)
codigos = {'SELIC': 432}

# Este comando coleta ambas as séries em um único DataFrame, pronto para a próxima etapa.
df_dados_brutos = sgs.get(
    codigos,
    start=data_inicial,
    end=data_final
).dropna()

df_dados_brutos.head() # Irá mostrar CDI e SELIC lado a lado
```

```python
df_dados_brutos['CDI'] = df_dados_brutos['SELIC'] - 0.10
```

sgs	É o módulo Sistema Gerenciador de Séries Temporais da biblioteca bcb.

.get()	É o comando de "obter" ou "buscar".

codigos	É o dicionário que criei: {'CDI': 12, 'SELIC': 432}.

(Ele informa à função quais séries temporais buscar (12 e 432) e como nomeá-las no DataFrame (CDI e SELIC).)

O .dropna() é o seu passo de limpeza inicial, assegurando que seu cálculo de juros compostos seja feito apenas em datas com dados válidos e completos.

```python
df_dados_brutos['CDI_Decimal'] = df_dados_brutos['CDI'] / 100
df_dados_brutos['SELIC_Decimal'] = df_dados_brutos['SELIC'] / 100
```

```python
df_dados_brutos
```

```python
# --- CÓDIGO CORRIGIDO PARA NORMALIZAR A UNIDADE ---

# Normaliza para garantir que a taxa ANUAL esteja em formato decimal (Ex: 0.1375)
df_dados_brutos['CDI_Decimal'] = df_dados_brutos['CDI'].apply(
    lambda x: x / 100 if x > 1 else x
)
df_dados_brutos['SELIC_Decimal'] = df_dados_brutos['SELIC'].apply(
    lambda x: x / 100 if x > 1 else x
)

# Agora, o cálculo diário (4ª imagem) será feito sobre a taxa correta.
df_dados_brutos['CDI_Diario'] = ((1 + df_dados_brutos['CDI_Decimal'])**(1/252)) - 1
# ... etc.
```

```python
df_dados_brutos['CDI_Decimal'] = df_dados_brutos['CDI'] / 100
df_dados_brutos['SELIC_Decimal'] = df_dados_brutos['SELIC'] / 100
```

```python
df_dados_brutos['CDI_Diario'] = ((1 + df_dados_brutos['CDI_Decimal'])**(1/252)) - 1
df_dados_brutos['SELIC_Diario'] = ((1 + df_dados_brutos['SELIC_Decimal'])**(1/252)) - 1
```

```python
# Exibir as colunas relevantes para verificar a conversão
print("\n--- Taxas Anuais vs. Taxas Diárias (Decimal) ---")
print(df_dados_brutos[['CDI', 'CDI_Diario', 'SELIC', 'SELIC_Diario']].head())
```

```python
df_dados_brutos
```

```python
# 1. Definir o valor inicial
VALOR_INICIAL = 1000.00

# 2. Criar as colunas de Saldo Inicial, preenchendo o primeiro valor com R$ 1000,00
# (Você só precisa criar estas colunas uma vez)
df_dados_brutos['Saldo_TS'] = VALOR_INICIAL # Tesouro Selic
df_dados_brutos['Saldo_CDB'] = VALOR_INICIAL # CDB / Caixinha Nubank/Mercado Pago
df_dados_brutos['Saldo_LCI'] = VALOR_INICIAL # LCI/LCA
df_dados_brutos['Saldo_CDB_120'] = VALOR_INICIAL
```

```python
# Iterar sobre as linhas do DataFrame a partir da segunda linha (índice 1)
# O índice 0 já tem o valor inicial (R$ 1000,00)
for i in range(1, len(df_dados_brutos)):
    # --- Tesouro Selic (100% da Selic Diária) ---
    taxa_ts_diaria = df_dados_brutos.iloc[i]['SELIC_Diario']
    saldo_ontem_ts = df_dados_brutos.iloc[i-1]['Saldo_TS']
    # O comando .iloc com get_loc garante que o valor seja salvo na célula correta
    df_dados_brutos.iloc[i, df_dados_brutos.columns.get_loc('Saldo_TS')] = saldo_ontem_ts * (1 + taxa_ts_diaria)

    # --- CDB / Caixinha Mercado Pago (100% do CDI Diário) ---
    taxa_cdb_diaria = df_dados_brutos.iloc[i]['CDI_Diario']
    saldo_ontem_cdb = df_dados_brutos.iloc[i-1]['Saldo_CDB']
    df_dados_brutos.iloc[i, df_dados_brutos.columns.get_loc('Saldo_CDB')] = saldo_ontem_cdb * (1 + taxa_cdb_diaria)

    # --- LCI / LCA (Ex. 90% do CDI Diário - Isento de IR) ---
    taxa_lci_diaria = df_dados_brutos.iloc[i]['CDI_Diario'] * 0.90
    saldo_ontem_lci = df_dados_brutos.iloc[i-1]['Saldo_LCI']
    df_dados_brutos.iloc[i, df_dados_brutos.columns.get_loc('Saldo_LCI')] = saldo_ontem_lci * (1 + taxa_lci_diaria)


    # ... Tesouro Selic (TS) ...
    # ... CDB 100% (CDB) ...

    # --- NOVO: CDB 120% do CDI ---
    taxa_cdb_120_diaria = df_dados_brutos.iloc[i]['CDI_Diario'] * 1.20
    saldo_ontem_cdb_120 = df_dados_brutos.iloc[i-1]['Saldo_CDB_120']

    # Atualiza o saldo: Saldo anterior * (1 + taxa)
    df_dados_brutos.iloc[i, df_dados_brutos.columns.get_loc('Saldo_CDB_120')] = saldo_ontem_cdb_120 * (1 + taxa_cdb_120_diaria)

# Exibir a evolução do Saldo Acumulado nos primeiros e últimos dias (Verificação)
print("\n--- Saldo Acumulado (Bruto) ---")
print(df_dados_brutos[['Saldo_TS', 'Saldo_CDB', 'Saldo_LCI', 'Saldo_CDB_120']].tail())
```

```python
# Definir o valor inicial para o cálculo do lucro
VALOR_INICIAL = 1000.00
ALIQUOTA_IR = 0.15 # 15% em decimal, para resgates acima de 2 anos (5 anos)
```

```python
# 1. Cálculo do Lucro Bruto Total (Saldo Final - Principal)
lucro_bruto_ts = df_dados_brutos['Saldo_TS'].iloc[-1] - VALOR_INICIAL
lucro_bruto_cdb = df_dados_brutos['Saldo_CDB'].iloc[-1] - VALOR_INICIAL
lucro_bruto_lci = df_dados_brutos['Saldo_LCI'].iloc[-1] - VALOR_INICIAL
lucro_bruto_cdb_120 = df_dados_brutos['Saldo_CDB_120'].iloc[-1] - VALOR_INICIAL
```

```python
lucro_bruto_cdb_120
```

```python
# 2. Cálculo do IR e Lucro Líquido
# Tesouro Selic e CDB (Pagam IR de 15% sobre o lucro)
ir_pago_ts = lucro_bruto_ts * ALIQUOTA_IR
lucro_liquido_ts = lucro_bruto_ts - ir_pago_ts
```

```python
ir_pago_cdb = lucro_bruto_cdb * ALIQUOTA_IR
lucro_liquido_cdb = lucro_bruto_cdb - ir_pago_cdb
```

```python
# LCI/LCA (Isento de IR)
ir_pago_lci = 0.00
lucro_liquido_lci = lucro_bruto_lci # Lucro Líquido = Lucro Bruto
```

```python
ir_pago_cdb_120 = lucro_bruto_cdb_120 * ALIQUOTA_IR
lucro_liquido_cdb_120 = lucro_bruto_cdb_120 - ir_pago_cdb_120
```

```python
# 3. Cálculo do Saldo Final Líquido
saldo_final_liquido_ts = VALOR_INICIAL + lucro_liquido_ts
saldo_final_liquido_cdb = VALOR_INICIAL + lucro_liquido_cdb
saldo_final_liquido_lci = VALOR_INICIAL + lucro_liquido_lci
```

```python
saldo_final_liquido_cdb_120 = VALOR_INICIAL + lucro_liquido_cdb_120
```

```python
# 1. Calcular o Fator de Desconto do IR (1 - Alíquota)
# Se paga 15% de IR, o fator líquido é 1 - 0.15 = 0.85 (ou seja, sobram 85% do lucro)
FATOR_LIQUIDO = 1 - ALIQUOTA_IR
```

```python
# Saldo Líquido Tesouro Selic
df_dados_brutos['Saldo_Liquido_TS'] = VALOR_INICIAL + (
    (df_dados_brutos['Saldo_TS'] - VALOR_INICIAL) * FATOR_LIQUIDO)
```

```python
# Saldo Líquido CDB
df_dados_brutos['Saldo_Liquido_CDB'] = VALOR_INICIAL + (
    (df_dados_brutos['Saldo_CDB'] - VALOR_INICIAL) * FATOR_LIQUIDO)
```

```python
# Saldo Líquido LCI/LCA (IR = 0%, então o Saldo Bruto é o Saldo Líquido)
df_dados_brutos['Saldo_Liquido_LCI'] = df_dados_brutos['Saldo_LCI']
```

```python
#Saldo Liquido CDB 120%
df_dados_brutos['Saldo_Liquido_CDB_120'] = VALOR_INICIAL + (
    (df_dados_brutos['Saldo_CDB_120'] - VALOR_INICIAL) * FATOR_LIQUIDO)
```

```python
# Exibir as últimas linhas para ver a curva líquida
print("\n--- Curvas de Crescimento Líquido (Pronto para o Power BI) ---")
print(df_dados_brutos[['Saldo_Liquido_TS', 'Saldo_Liquido_CDB', 'Saldo_Liquido_LCI', 'Saldo_Liquido_CDB_120']].tail())
```
